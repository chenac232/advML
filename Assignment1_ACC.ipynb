## 1. Explore Bivariate results
Describe any relationships you see between particular features and the target variable

- We'll look at GDP per capita, perceptions of corruption and happiness.

```python
import pandas as pd

# Merge y_train and x_train to look at relationship between specific features and the target variable
df = pd.merge(X_train, y_train, right_index=True, left_index=True)

# recode y_train column to a "happiness" column
df['happy'] = 0
df.loc[(df['Very Low']==1), 'happy'] = 1
df.loc[(df['Low']==1), 'happy'] = 2
df.loc[(df['Average']==1), 'happy'] = 3
df.loc[(df['High']==1), 'happy'] = 4
df.loc[(df['Very High']==1), 'happy'] = 5
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# create box plot for features
f2, ax = plt.subplots(2, 2, figsize=(16, 12))
sns.boxplot('happy', 'GDP per capita', data=df, ax=ax[0, 0], palette='Blues')
sns.boxplot('happy', 'Perceptions of corruption', data=df, ax=ax[0, 1], palette='Blues')
sns.boxplot('happy', 'Terrorist_attacks', data=df, ax=ax[1, 0], palette='Blues')
sns.boxplot('happy', 'Social support', data=df, ax=ax[1, 1], palette='Blues')
```

We can see from the boxplots above that GDP per capita has the most linear relationship with happiness. The greater GDP per capita, generally, the greater happiness. For other features, it's a little less linear to determine. Perceptions of corruption and terrorist attacks do not appear to have a linear relationship with how happy people are. Social support is a relatively strong indicator of happiness. Countries with greater social support tend to have greater happiness.

## 1. PCA

```python
# PCA implementation
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA

numeric_features = df.drop(['Country or region', 'name', 'region', 'sub-region'], axis=1)
# numeric_features=numeric_features.columns.tolist()
```

```python
numeric_features.head(5)
```

```python


df = numeric_features
X = pd.DataFrame(scale(df), index=df.index, columns=df.columns)

# fit PCA with 2 components
pca = PCA(n_components=2)
pca.fit_transform(X)
```

```python
# The loading vectors
pca_loadings = pd.DataFrame(PCA(n_components=2).fit(X).components_.T, index=df.columns)
pca_loadings
```

From the table above, we can see that GDP per capita, social support, and healthy life expectancy tend to be strong determinents of happiness relative to the other variables present.

```python
print(pca.components_) 
```

```python
print(pca.explained_variance_) # explained variance for for each principal component vector
```

```python
pca.explained_variance_ratio_
```

First PCA explains about 40 percent of the variance, second PCA explains 16 percent of the variance

### Set up

```python
# Set up preprocessing

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# We create the preprocessing pipelines for both numeric and categorical data.

numeric_features = X_train.drop(['Country or region', 'name', 'region', 'sub-region'], axis=1)
numeric_features=numeric_features.columns.tolist()

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

categorical_features = ['region', 'sub-region']

#Replacing missing values with Modal value and then one hot encoding.
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# final preprocessor object set up with ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

#Fit your preprocessor object
preprocess=preprocessor.fit(X_train) 
```

```python
# Write function to transform data with preprocessor

def preprocessor(data):
    data.drop(['Country or region', 'name'], axis=1)
    preprocessed_data=preprocess.transform(data)
    return preprocessed_data
```

```python
# check shape of X data after preprocessing it using our new function
preprocessor(X_train).shape
```
